Operações Básicas em Python

```python
# Adição
# Subtração
# Divisão
# Módulo
# Potência

print("Resultado da adição: " + str(2+2)) # Em Python quando for concatenar string com número, tem que converter tudo em string
print("Resultado da subtração: " + str(2-2))
print("Resultado da divisão: " + str(2/2))
print("Resultado do módulo: " + str(2%2))
print("Resultado do potenciação: " + str(2**2))
```
