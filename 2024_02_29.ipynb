Programação em Python Set e Frozenset

Set - é uma lista mutável
Frozenset - é uma lista imutável

```python
# Criando um set

lista_set = {1,2,3,'texto',(1,2,3)}
print(lista_set)
```

```python
# Com sequências variáveis

x = set('Francisco')
print(x)

list(x)
```

```python
# Frozenset
# A estrutura não possui nenhuma função de manipulação

y = set()
y.add(1)
y.add(2)

print(len(y))
```

```python
# Operações de conjuntos
# | união
# & intersecção
# - diferença

a = {1,2,3,4}
b = {4,5,6,7,8}

print(a|b)
print(a&b)
print(a-b)
```

```python
# Dicionários
# chave e valor
# {a:1, b:2, c:3, ...}

# Chaves = objetos imutáveis
dict = {1:[1,[1,2]], 2:'Texto qualquer'}

# Selecionar um valor no dicionário
dict2 = {'a':1*10 ** 1, 'b':(1,1), 'c':9}
print(dict2['a'])
print(dict2['b'])
print(dict2['c'])

print(dict2.get('b'))
```

```python
# Algumas funções nos dicionários
# popitem() - retorna o último valor
# pop(chave) - retorna o valor da chave
# update({chave:valor}) - altera o valor da chave

dict3 = {1:1*10**1, 2:2*10**2, 3: 3*10**3, 4:'texto', 'p':2000, 5:90, 'palavra':'outro texto'}
print(dict3.get(1))
print(dict3.get(4))
print(dict3.get(4))
print(dict3.get('palavra'))
print(dict3.get('p'))
```

```python
# Estrutura de decisão - if, elif, else

isEstudante = False
isGravida = False
idade = 84
multiplicador_desconto = 0.45
isIdoso = True

valor_cinema = 10.00
valor_final = valor_cinema

if isEstudante == True:
  valor_final = valor_final * 0.5
elif idade >= 85:
  valor_final = valor_final * 0.7
elif isGravida == True:
  valor_final = valor_final * 0.75
else:
  valor_final = valor_final * multiplicador_desconto

print(f'O valor do ingresso é {valor_final}')
```

```python
# Como criar uma função
# public void nome_funcao(){}
# public function nome_funcao(){}

# def nome_funcao():
#   [escopo da função]

def preco_do_cinema(preco, multiplicador):
  return preco * multiplicador

preco = 10.00
multiplicador = 0.5

isIdoso = False
idade = 86

if idade >= 86:
  multiplicador = 0.4
elif isIdoso == True:
  multiplicador = 0.7

print(f"O valor do ingresso é R${preco_do_cinema(preco, multiplicador)}")
```

```python
def listar_mes(mes, dia):
  nome=""
  texto=""

  if mes == 1:
    #print(nome)
    nome="Roberto"
  elif mes == 2:
    #print(nome2)
    nome="Ana"

  if dia == 10:
    #print(f'Aniversário do {nome}')
    texto = 'Aniversário do'
  elif dia == 16:
    #print(f'Aniversário da')
    texto = 'Aniversário da'

  return texto, nome
```

```python
#print(listar_mes(2,16))
resTexto, resNome = listar_mes(2,16)

print(resTexto)
print(resNome)
```
