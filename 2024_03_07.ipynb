```python
# Laço de repetição
# for

i=0
retorno = ""

for letra in "Roberto":
  retorno += letra
  i+=1

  if i > 2:
    print(retorno)
    retorno = ""
    i=0
  else:
    print(letra)
```

```python
nomes = ["João", "Antônio", "Roberto", "Carlos"]
aniversarios = ["15/12/1995", "13/01/1990", "31/12/2000", "20/10/2013"]

nomes_aniversarios = list(zip(nomes, aniversarios))

for nome, aniversario in nomes_aniversarios:
  print(f"O aniversário de {nome} é dia {aniversario}")
```

```python
# Exemplo break
# Exemplo continue

banido = "Roberto"

for nome in nomes:
  if banido == nome:
    print(f"O cliente {banido} não entrou na sala de jogos")
    # break
    continue

  print(f"O cliente {nome} entrou na sala de jogos")
```

```python
from random import shuffle, random

clientes = ["José", "Marcos", "Antônio"]
produtos = ["Frango a Passarinho", "Ovo frito", "Costelinha frita", "Queijo de nózinho"]

alegria = "Frango a Passarinho"

for cliente in clientes:
  print(f"O {cliente} entrou no restaurante")

  shuffle(produtos)

  for produto in produtos:
    print(f"O {cliente} consumiu um(a) {produto}")

    if produto == alegria:
      print("O cliente saiu satisfeito")
      break

    if (random() * 10) > 5:
      print(f"{cliente} bate um papo")
    else:
      print(f"o {cliente} pede para tocar uma sofrência")

  print("O cliente foi embora")
  print("")
```
