```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('/content/drive/MyDrive/Sistemas para Internet/Tópicos em Computação/Curso_Python/imigrantes.csv')
```

```python
data
```

```python
data.info()
```

```python
data.set_index('Country', inplace=True)
```

```python
anos = list(map(str, range(1980, 2014)))
```

```python
anos
```

```python
imigracao_brasil = data.loc['Brazil',anos]
```

```python
imigracao_brasil
```

```python
dicionario = {'ano':imigracao_brasil.index.tolist(), 'total':imigracao_brasil.values.tolist()}
```

```python
dados = pd.DataFrame(dicionario)
```

```python
dados.head()
```

```python
plt.plot(dados['ano'], dados['total'])
```

```python
plt.plot(dados['ano'], dados['total'])
plt.xticks(['1980', '1985', '1990', '1995', '2000', '2005', '2010'])
```

```python
plt.plot(dados['ano'], dados['total'])
plt.xticks(['1980', '1985', '1990', '1995', '2000', '2005', '2010'])
plt.yticks([500, 1000, 1500, 2000, 2500, 3000])
plt.show()
```

```python
plt.figure(figsize=(10,7))
plt.plot(dados['ano'], dados['total'])
plt.xticks(['1980', '1985', '1990', '1995', '2000', '2005', '2010'])
plt.yticks([500, 1000, 1500, 2000, 2500, 3000])
plt.show()
```

```python
plt.figure(figsize=(10,7))
plt.plot(dados['ano'], dados['total'])
plt.title('Total de Imigração do Brasil para o Canadá')
plt.xticks(['1980', '1985', '1990', '1995', '2000', '2005', '2010'])
plt.yticks([500, 1000, 1500, 2000, 2500, 3000])
plt.show()
```

```python
plt.figure(figsize=(10,7))
plt.plot(dados['ano'], dados['total'])
plt.title('Total de Imigração do Brasil para o Canadá')
plt.xlabel('Ano')
plt.ylabel('Número de imigrantes')
plt.xticks(['1980', '1985', '1990', '1995', '2000', '2005', '2010'])
plt.yticks([500, 1000, 1500, 2000, 2500, 3000])
plt.show()
```

```python
fig, ax = plt.subplots(figsize=(7,3))
ax.plot(dados['ano'], dados['total'])
plt.show()
```

```python
fig, ax = plt.subplots(figsize=(7,3))
ax.plot(dados['ano'], dados['total'])
ax.xaxis.set_major_locator(plt.MultipleLocator(5))
plt.show()
```

```python
fig, ax = plt.subplots(figsize=(7,3))
ax.plot(dados['ano'], dados['total'])
ax.set_title('Imigração do Brasil para o Canadá')
ax.set_ylabel('Número de imigrantes')
ax.set_xlabel('Ano')
ax.xaxis.set_major_locator(plt.MultipleLocator(5))
plt.show()
```
