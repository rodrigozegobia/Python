```python
fig, axis = plt.subplots(2,2, figsize=(10,6))

# Ajusta o espaçamento entre os gráficos
fig.subplots_adjust(hspace=0.5, wspace=0.3)

# Título
fig.suptitle('Imigração dos quatro maiores países da América do Sul para o Canadá', fontsize=16)

# Subplots
axis[0,0].plot(data.loc['Brazil', anos], lw=2, marker='.', color='g')
axis[0,0].set_title('Brazil')

axis[0,1].plot(data.loc['Colombia', anos], lw=2, color='r')
axis[0,1].set_title('Colômbia')

axis[1,0].plot(data.loc['Argentina', anos], lw=2, color='b')
axis[1,0].set_title('Argentina')

axis[1,1].plot(data.loc['Peru', anos], lw=2, color='y')
axis[1,1].set_title('Peru')

# Alterando a frequência dos ticks do eixo x
# Também altera os labels
for ax in axis.flat:
  ax.xaxis.set_major_locator(plt.MultipleLocator(5))
  ax.set_xlabel('Ano', fontsize=14)
  ax.set_ylabel('Número de imigrantes', fontsize=12)
  '''
  ax.yaxis.set_tick_params(labelsize=12)
  ax.xaxis.set_tick_params(labelsize=12)
  '''
  ax.grid(linestyle='--')

# Alterando o máximo e mínimo do gráfico
ymin=0
ymax=7000

for ax in axis.ravel():
  ax.set_ylim(ymin, ymax)

plt.show()
```

```python
paises_america_sul = data.query('Region == "South America"')
paises_america_sul.head()
```

```python
colors = ['royalblue', 'orange', 'forestgreen', 'orchid', 'purple', 'brown', 'slateblue', 'gray', 'olive', 'navy', 'teal', 'tomato']

fig, axis = plt.subplots(figsize=(12,5))

axis.barh(paises_america_sul.index, paises_america_sul['Total'], color=colors)
axis.set_title('Imigração da América do Sul para o Canadá')
axis.set_ylabel('Número de imigrantes', fontsize=14)
axis.set_xlabel('')
axis.yaxis.set_tick_params(labelsize=12)
axis.xaxis.set_tick_params(labelsize=12)

plt.show()
```

```python
paises_america_sul_sorted = paises_america_sul.sort_values(by='Total', ascending=True)
paises_america_sul_sorted
```

```python
colors = ['royalblue', 'orange', 'forestgreen', 'orchid', 'purple', 'brown', 'slateblue', 'gray', 'olive', 'navy', 'teal', 'tomato']

fig, axis = plt.subplots(figsize=(12,5))

axis.barh(paises_america_sul_sorted.index, paises_america_sul_sorted['Total'], color=colors)
axis.set_title('Imigração da América do Sul para o Canadá')
axis.set_ylabel('Número de imigrantes', fontsize=14)
axis.set_xlabel('')
axis.yaxis.set_tick_params(labelsize=12)
axis.xaxis.set_tick_params(labelsize=12)

plt.show()
```

```python
colors = []

for pais in paises_america_sul_sorted.index:
  if pais == 'Brazil':
    colors.append('green')
  else:
   colors.append('silver')

fig, axis = plt.subplots(figsize=(12,5))

axis.barh(paises_america_sul_sorted.index, paises_america_sul_sorted['Total'], color=colors)
axis.set_title('Imigração da América do Sul para o Canadá')
axis.set_ylabel('Número de imigrantes', fontsize=14)
axis.set_xlabel('')
axis.yaxis.set_tick_params(labelsize=12)
axis.xaxis.set_tick_params(labelsize=12)

# Loop para adicionar os rótulos em cada barra
for i, v in enumerate(paises_america_sul_sorted['Total']):
  axis.text(v + 20, i, str(v), color='black', fontsize='9', ha='left', va='center')

axis.set_frame_on(False)

axis.get_xaxis().set_visible(False)
axis.tick_params(axis='both', which='both', length=0)

fig.savefig(path+'imigracao_america_canada.jpg', transparent=False, bbox_inches='tight')

plt.show()
```

```python
# https://seaborn.pydata.org/tutorial/color_palettes.html

def criar_grafico_com_paleta(paleta):
  fig, ax = plt.subplots(figsize=(8,4));
  ax=sns.barplot(data=dez_paises, y=dez_paises.index, x='Total', orient='h', palette=paleta, hue=dez_paises.index);
  ax.set_title('Países que tem a maior imigração para o Canadá');
  ax.set_xlabel('Total de imigrantes');
  ax.set_ylabel('');
```

```python
criar_grafico_com_paleta('Blues');
```

```python
criar_grafico_com_paleta('Blues_r');
```

```python
criar_grafico_com_paleta('rocket');
```

```python
criar_grafico_com_paleta('coolwarm');
```

```python
criar_grafico_com_paleta("tab10");
```

```python
sns.set_theme(style="whitegrid")
criar_grafico_com_paleta("tab10");
```

```python
sns.set_theme(style="ticks")
criar_grafico_com_paleta("tab10");
```
