```python
dados1.head(10)
```

```python
dados1['RACA'] = dados1['RACA'].replace({0:'na_declarado', 1: 'branca', 2: 'preta', 3: 'parda', 4: 'amarela', 5: 'indígena'})
```

```python
dados1.head()
```

```python
dados1['TP_LINGUA'] = dados1['TP_LINGUA'].replace([0,1],['Ingles', 'Espanhol'])
```

```python
dados1['ESCOLA'] = dados1['ESCOLA'].replace([1,2,3,4],['não_respondeu','publica', 'privada', 'exterior'])
```

```python
dados1['INTERNET'] = dados1['INTERNET'].replace(['A','B'],['Não', 'Sim'])
```
